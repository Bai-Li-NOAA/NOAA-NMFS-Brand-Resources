# NMFS Icon Maker Script

```python
from PIL import Image

def resize_image(input_path, output_path, size):
    """
    Resize an image to a specific size.
    
    Args:
    - input_path (str): Path to the input image file.
    - output_path (str): Path to save the resized image.
    - size (tuple): Target size in pixels (width, height).
    """
    with Image.open(input_path) as img:
        img_resized = img.resize(size, Image.LANCZOS)
        img_resized.save(output_path)

def generate_favicon(input_path, output_path):
    """
    Generate a favicon.ico file from a single image.
    
    Args:
    - input_path (str): Path to the input image file (16x16 icon).
    - output_path (str): Path to save the favicon.ico file.
    """
    # Resize the image to 16x16
    resized_image = Image.open(input_path)
    resized_image = resized_image.resize((16, 16), Image.LANCZOS)
    
    # Save the resized image as favicon.ico
    resized_image.save(output_path, format="ICO")

# Path to the input PNG image
input_image_path = "./noaa_digital_logo-2022_icon.png"

# List of desired sizes for icons
icon_sizes = [
    (16, 16),
    (32, 32),
    (57, 57),
    (60, 60),
    (72, 72),
    (76, 76),
    (114, 114),
    (120, 120),
    (144, 144),
    (152, 152),
    (180, 180)
]

# Loop through each size and resize the image
for size in icon_sizes:
    output_path = f"noaa-icon-{size[0]}x{size[1]}.png"
    resize_image(input_image_path, output_path, size)

# Path to the input PNG image (16x16 icon)
icon_image_path = "./noaa-icon-16x16.png"

# Generate favicon.ico file
generate_favicon(icon_image_path, "favicon.ico")
from IPython.display import Image, display

# Display each PNG file
for size in icon_sizes:
    output_path = f"noaa-icon-{size[0]}x{size[1]}.png"
    display(Image(filename=output_path))

```
